```python
#Import Libraries
import pandas as pd
import numpy as np
import plotly.express as px
import datetime
from google.colab import files
```

```python
#Import data
df = pd.read_csv("https://raw.githubusercontent.com/aakap/NYC_factors-that-affect-biking/main/midtermprojectrawdata.csv")
df.head(100)

```

```python
#data cleaning
df["Precipitation"].replace({"T": "0.00"}, inplace=True);
df['Weekday'] = df["Day"].replace({"Saturday":"1", "Sunday":"2", "Monday":"3","Tuesday":"4","Wednesday":"5", "Thursday":"6","Friday":"7"})
df = df.drop(columns=['Date', 'Day'])
df.head(100)
```

```python
#graphs 

#scatterplot of high temp vs total crossings
hightemp = px.scatter(df, x = "High Temp (°F)",
                            y = "Total",
                            title = "Total Crossings vs High Temperatures",)
hightemp
```

```python
#scatterplot of low temp vs total crossings
lowtemp = px.scatter(df, x = "Low Temp (°F)",
                            y = "Total",
                            title = "Total Crossings vs Low Temperatures",)
lowtemp
```

```python
#scatterplot of precipitation vs total crossings
precipitation = px.scatter(df, x = "Precipitation",
                            y = "Total",
                            title = "Total Crossings vs Precipitation")
precipitation
#scatterplot of day of week vs total crossings
```

```python
#scatterplot of day of week vs total crossings
dayofweekgraph = px.scatter(df, x = "Weekday",
                            y = "Total",
                            title = "Total Crossings vs Weekday", labels={"1":"Saturday"})
dayofweekgraph
```

```python
import statsmodels.api as sm # import statsmodels 
import numpy as np
df1 = df.astype(np.float)

#df = np.asarray(df)
X = df1[["High Temp (°F)","Low Temp (°F)","Precipitation","Weekday"]] #initialize independent variables
y = df1["Total"] #initialize dependent variable
model = sm.OLS(y, X).fit() #define least of squares regression model
predictions = model.predict(X) #run model
model.summary() #print results 
```
